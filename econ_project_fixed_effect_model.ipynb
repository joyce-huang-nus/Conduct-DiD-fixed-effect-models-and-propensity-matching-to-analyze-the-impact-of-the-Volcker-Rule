```python
import pandas as pd
import numpy as np

from statsmodels.formula.api import ols
```

```python
# Data cleaning - delete rows that have null values
df = pd.read_csv(r'C:\Users\Joyce Huang\Downloads\econ_group_project-02\DiD_data.csv')
df.head()
df = df.dropna() 
print(df.count())
df

print(df.groupby(['rssd9001'])['bhc_avgtradingratio'].count())
```

```python
# DiD baseline model
regout = ols('bhc_avgtradingratio ~ after_DFA_1*treat_3_b_avg + treat_3_b_avg + after_DFA_1 + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter', df).fit()
regout.summary2()
```

```python
# DiD fixed-bhc-effect model-data-preparation
pd.get_dummies(df['rssd9001'])
x = pd.get_dummies(df['rssd9001'])
df = pd.concat([x, df], axis=1)
df
```

```python
# DiD fixed-bhc-effect model
regout = ols('bhc_avgtradingratio ~ after_DFA_1*treat_3_b_avg + treat_3_b_avg + after_DFA_1 + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter + x', df).fit()
regout.summary2()
```

```python
# DiD fixed-time-effect model-data-preparation
print(pd.get_dummies(df['rssd9999']))
time = pd.get_dummies(df['rssd9999'])
df = pd.concat([time, df], axis=1)
df
```

```python
# DiD fixed-time-effect model
regout = ols('bhc_avgtradingratio ~ after_DFA_1*treat_3_b_avg + treat_3_b_avg + after_DFA_1 + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter + time', df).fit()
regout.summary2()
```

```python
# DiD fixed-effect model
regout = ols('bhc_avgtradingratio ~ after_DFA_1*treat_3_b_avg + treat_3_b_avg + after_DFA_1 + dep_roa1 + dep_leverage + dep_lnassets + dep_creditrisk_total3 + dep_cir + dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter + time + x', df).fit()
regout.summary2()
```
