## robustness tests: matching

```python
df2 = df[df['rssd9999']=='20040930']
df2.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
model = lr_model.fit(df2.iloc[:,5:14], df2.iloc[:,3])
y_predicted_proba = model.predict_proba(df2.iloc[:,5:14]) 
df2['score'] = y_predicted_proba[:,1]
```

```python
treats = df2[df2['treat_3_b_avg'] == 1].iloc[:,0]
controls = df2[df2['treat_3_b_avg'] == 0].iloc[:,0]
data = pd.DataFrame(columns = [treats.values])
data.insert(0,'control',controls.values)
```

```python
for i in range(data.shape[0]):
    for j in range(1,data.shape[1]):
        data.iloc[i,j] = abs(df2[df2['rssd9001']==data.iloc[i,0]]['score'].values-df2[df2['rssd9001']==treats.values[j-1]]['score'].values)
```

```python
data
```

```python
match = np.empty([data.shape[1]-1,3],dtype=object)
for i in range(match.shape[0]):
    match[i]=data.sort_values(by=data.columns[i+1]).iloc[0:3,0].values
match
```

```python
match_data = pd.DataFrame(columns=df.columns.values)
treat_data = df[df['treat_3_b_avg'] == 1]
match_data = match_data.append(treat_data,ignore_index=True,sort=False)
for i in range(match.shape[0]):
    for j in range(match.shape[1]):
        newdata = df[df['rssd9001']== match[i,j]]
        match_data = match_data.append(newdata,ignore_index=True,sort=False)
match_data
```

```python
regout = ols('bhc_avgtradingratio ~ treat_3_b_avg + after_DFA_1+ treat_3_b_avg*after_DFA_1+rssd9999\
                     +rssd9001+dep_roa1 + dep_leverage+dep_lnassets + dep_creditrisk_total3 + dep_cir +\
                     dep_depositratio + dep_loans_REratio + dep_liquidity + dep_cpp_bankquarter', match_data).fit()
regout.summary2()
```
